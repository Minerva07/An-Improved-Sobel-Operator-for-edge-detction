```python
import random 
import numpy as np
import cv2 
```

```python
def add_noise(img): 
  
    # Getting the dimensions of the image 
    row , col = img.shape 
      
    # Randomly pick some pixels in the 
    # image for coloring them white 
    # Pick a random number between 300 and 10000 
    number_of_pixels = random.randint(300, 10000) 
    for i in range(number_of_pixels): 
        
        # Pick a random y coordinate 
        y_coord=random.randint(0, row - 1) 
          
        # Pick a random x coordinate 
        x_coord=random.randint(0, col - 1) 
          
        # Color that pixel to white 
        img[y_coord][x_coord] = 255
          
    # Randomly pick some pixels in 
    # the image for coloring them black 
    # Pick a random number between 300 and 10000 
    number_of_pixels = random.randint(300 , 10000) 
    for i in range(number_of_pixels): 
        
        # Pick a random y coordinate 
        y_coord=random.randint(0, row - 1) 
          
        # Pick a random x coordinate 
        x_coord=random.randint(0, col - 1) 
          
        # Color that pixel to black 
        img[y_coord][x_coord] = 0
          
    return img 
  
# salt-and-pepper noise can be applied only to grayscale images 
# Reading the color image in grayscale image 
img = cv2.imread("C:\\Users\\marys\\OneDrive\\Documents\\SPML_sem_1\\Sp_Lab\\Capstone project\\Lena.png", cv2.IMREAD_GRAYSCALE) 

#Display the original image
cv2.imshow('Original image', img)   

noisy_img =  add_noise(img)
    
#Storing the noisy image 
cv2.imshow('Noisy image', noisy_img)
```

```python
def sobel_edge_detection(image):
    # Define Sobel operators for horizontal and vertical edges
    sobel_x = np.array([[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]])
    sobel_y = np.array([[-1, -2, -1], [0, 0, 0], [1, 2, 1]])

    # Get the dimensions of the image
    height, width = image.shape

    # Create empty arrays to store the horizontal and vertical gradients
    gradient_x = np.zeros((height, width))
    gradient_y = np.zeros((height, width))

    # Perform convolution to calculate the gradients
    for i in range(1, height - 1):
        for j in range(1, width - 1):
            window = image[i - 1:i + 2, j - 1:j + 2]
            gradient_x[i, j] = np.sum(sobel_x * window)
            gradient_y[i, j] = np.sum(sobel_y * window)

    # Calculate the magnitude of the gradient
    gradient_magnitude = np.sqrt(gradient_x**2 + gradient_y**2)

    # Normalize the gradient magnitude to the range [0, 255]
    gradient_magnitude = (gradient_magnitude / np.max(gradient_magnitude)) * 255

    # Convert the result to 8-bit for display
    edge_image = np.uint8(gradient_magnitude)

    return edge_image

# Load the image
image = cv2.imread("C:\\Users\\marys\\OneDrive\\Documents\\SPML_sem_1\\Sp_Lab\\Capstone project\\Lena.png", cv2.IMREAD_GRAYSCALE)

# Perform Sobel edge detection
edge_image = sobel_edge_detection(img2)

# Display the edge-detected image

cv2.imshow('Edge Detected Image (Sobel)', edge_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
